```python
# TODO

import pandas as pd

# This CSV file contains semicolons instead of comas as separator
ds = pd.read_csv('assets/real_estate.csv', sep=';')
num_filas = ds.shape[0]
num_columnas = ds.shape[1]

salida = ""
for i in range(num_columnas):
    for j in range (num_filas):
        if pd.isna(ds.iloc[j, i]):
            salida = salida + f"{[i,j]}, "
      
if salida =="":
    print ("False")
else:
    print ("true"+ salida)
```

```python
from ipyleaflet import Map, basemaps, Marker, Icon, AwesomeIcon
import pandas as pd

# Map centered on (60 degrees latitude and -2.2 degrees longitude)
# Latitude, longitude


# Imprimir las capas base disponibles
#for basemap in basemaps.keys():
    #print(basemap)

map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenTopoMap)
#map

#He cargado un mapa diferente de los que estaban diponibles en la bilblioteca porque Stamen.Terrain no funcionaba.
```

```python
## Aquí: traza la coordenadas de los estados

## PON TU CÓDIGO AQUÍ:
diccionario_ej21= ds_filtradoej_13.set_index('Unnamed: 0')[['latitude','longitude','level5']].to_dict(orient = 'index')

small_icon = AwesomeIcon(icon='info-sign', icon_color='red', icon_size=(1, 1))
small_icon_2 = AwesomeIcon(icon='info-sign', icon_color='green', icon_size=(5, 5))
small_icon_3 = AwesomeIcon(icon='info-sign', icon_color='black', icon_size=(5, 5))
small_icon_4 = AwesomeIcon(icon='info-sign', icon_color='white', icon_size=(5, 5))

for Unnamed, valores in diccionario_ej21.items():
    latitude = float(valores['latitude'].replace(',', '.'))
    longitude = float(valores['longitude'].replace(',', '.'))
    level5 = valores['level5']  
    if level5 == 'Alcorcón':
        marcador = Marker(location=(latitude, longitude), icon = small_icon, draggable=True)
        map.add_layer(marcador)
    elif level5== 'Getafe':
        marcador = Marker(location=(latitude, longitude), icon = small_icon_2, draggable=True)
        map.add_layer(marcador)
    elif level5 == 'Fuenlabrada':
        marcador = Marker(location=(latitude, longitude), icon = small_icon_3, draggable=True)
        map.add_layer(marcador)
    elif level5 == 'Leganés':
        marcador = Marker(location=(latitude, longitude), icon = small_icon_4, draggable=True)
        map.add_layer(marcador)
        

    

map    
```
